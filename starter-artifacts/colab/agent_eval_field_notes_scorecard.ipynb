```python
import pandas as pd

url = 'https://raw.githubusercontent.com/MukundaKatta/agent-eval-platform-starters/main/starter-artifacts/dataset/agent_eval_field_notes_scenarios.csv'
df = pd.read_csv(url)
df.head()
```

```python
weights = {
    'task_success': 0.35,
    'tool_reliability': 0.25,
    'recovery_behavior': 0.20,
    'evidence_quality': 0.20,
}

sample = {
    'task_success': 8.2,
    'tool_reliability': 7.4,
    'recovery_behavior': 6.8,
    'evidence_quality': 7.6,
}

score = sum(sample[k] * weights[k] for k in weights)
readiness = 'ready' if score >= 8 else 'needs targeted review' if score >= 6.5 else 'not ready'
{'score': round(score, 2), 'readiness': readiness}
```

## How to use this

Use this notebook as a compact review surface. Replace the sample values with scores from a real agent trace, then keep the notes with the trace evidence that supports the score.
